# Importe das bibliotecas e dataframe

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
df = pd.read_csv('ProdutosMaisVendidos.csv')
```

```python
df.head()
```

# MudanÃ§as no dataframe e criaÃ§Ã£o da matrix de correlaÃ§Ã£o

```python
df['product_id2'] = df['product_id']
df_final = pd.get_dummies(df, columns=['product_id2'])
```

```python
df_final.head()
```

```python
cosseno = cosine_similarity(df_final.drop('product_id', axis=1).T)
index_df = df_final['product_id'].T.index
```

# Teste do modelo de recomendaÃ§Ã£o com base na ordem dos produtos mais bem avaliados, dentro dos 1000 produtos mais vendidos

```python
item_index = int(input('Digite a ordem do produto mais bem avaliado: '))


if item_index not in index_df:
    print('Produto nÃ£o encontrado, tente novamente')
closest_n_iten = np.argsort(cosseno[item_index])
itens_proximos = []
idx = len(closest_n_iten) - 1
while (idx >= 0):
    itens_proximos.append(closest_n_iten[idx])
    idx = idx - 1
itens_proximos = itens_proximos[1:11]
n_closest = []
while len(n_closest) < 10:
    for item_data in zip(df['product_id'][index_df[itens_proximos]], cosseno[item_index][itens_proximos]):
        n_closest.append(item_data)
tuple(n_closest)
```
